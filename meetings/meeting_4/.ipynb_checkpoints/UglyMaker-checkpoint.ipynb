```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
```

```python
full_data = pd.read_csv('data/cardio_train.csv', delimiter=";")
```

```python
messy_data = full_data[['age', 'height', 'weight', 'ap_hi', 'ap_lo', 'cardio']]
```

### Making it categorical

```python
full_data.head()
```

```python
def encode_chol(number: int):
    if number == 1:
        return "normal"
    elif number == 2:
        return "above normal"
    else:
        return "high above normal"
```

```python
messy_data['cholesterol'] = full_data.loc[:, 'cholesterol'].apply(encode_chol)
messy_data['glucose'] = full_data.loc[:, 'gluc'].apply(encode_chol)
messy_data
```

```python
def some_func(row: list):
    final_strings = []
    if row[0] == 1:
        final_strings.append("smokes")
    if row[1] == 1:
        final_strings.append("drinks")
    if row[2] == 1:
        final_strings.append("is_active")
        
        
    if len(final_strings) == 0:
        return np.nan
    else:
        return ", ".join(final_strings)
```

```python
messy_data['lifestyle'] = full_data[['smoke', 'alco', 'active']].apply(some_func, axis=1)
```
